### Estructura Básica en Pandas

Pandas ofrece dos tipos de clases para el manejo de los datos:

- **Series:** Una matriz unidimensional que puede contener todo tipo de datos como enteros, cadenas, objetos, etc.

- **DataFrame:** Se trata de una estructura bidimensional que contiene los datos dentro de una tabla con filas y columnas.

## Merge

### Concatenar

```python
df = pd.DataFrame(np.random.randn(10, 4))

df

# se crea un dataframe con numeros aleatorios en un rango de 10 filas y 4 columnas.
```

```python
piezas = [df[:3], df[3:7], df[7:]]
pd.concat(piezas)

# se crea una lista seccionando el dataframe en tres partes, luego con .concat se concatenan dichas secciones.
```

### Join

```python
left = pd.DataFrame({"key": ['Amoxicilina', 'Amoxicilina'], "lval": [60, 20]})
left

# En la variable left se crea una llave que sirve para definir el nombre y lval sirve para definir el valor de esa llave.
# al llamar a la variable salen las llaves y sus respectivos valores.
```

```python
right = pd.DataFrame({"key": ['Amoxicilina', 'Amoxicilina'], "rval": [40, 50]})
right

# Lo mismo ocurre con right solo que en esta ocasion la llave puede tener el mismo atributo o valor
# es decir, no se trata de una llave unica.
```

```python
pd.merge(left, right, on="key")

# Con .merge realizamos una fusion de ambas variables, obteniendo asi sus llaves y valores.
```

```python
left = pd.DataFrame({"key": ['Ibuprofeno', 'Amoxicilina'], "lval": [60, 20]})
right = pd.DataFrame({"key": ['Ibuprofeno', 'Amoxicilina'], "rval": [40, 50]})

pd.merge(left, right, on="key")

# .merge hace que que se combinen ambas variables integrando los valores a sus respectivas llaves.
```

## Agrupacion

```python
df = pd.DataFrame(
    {
        "A": ['Ambroxol', 'Alendronato', 'Omeprazol', 'Loratadina', 'Ibuprofeno', 'Amoxicilina'],
        "B": ["one", "one", "two", "three", "two", "three"],
        "C": np.random.randn(6),
        "D": np.random.randn(6),
    }
)
df

# Se crea un dataframe con varias columnas y estas albergan distintos tipos de valores.
```

```python
df.groupby("A")[["C", "D"]].sum()

# Se agrupan las filas de acuerdo a los valores de la columna "A", seguido de la suma de los valores de las columnas "C" y "D"
# segun el valor de "A". 
```

```python
df.groupby(["A", "B"]).sum()

# Se argupan las columnas "A" y "B" y se suman los valores originales de las columnas anteriores.
```

## Reshaping

```python
arrays = [
   ['Ambroxol', 'Alendronato', 'Omeprazol', 'Loratadina', 'Ibuprofeno', 'Amoxicilina'],
   ["one", "one", "two", "three", "two", "three"],
]

index = pd.MultiIndex.from_arrays(arrays, names=["first", "second"])
df = pd.DataFrame(np.random.randn(6, 2), index=index, columns=["A", "B"])
df2 = df[:4]
df2

# Se crean dos indices con los valores de medicamentos y numeros, seguido de la creacion de un avariable llamada index
# donde se nombran los encabezados de ambos indices como first and second, luego se crea la variable df en donde
# se crean valores aleatorios en un rango de 6 filas y dos columnas que esten relacionadas con la variable index y 
# que estas columnas lleven como nombre "A" y "B"
```

```python
stacked = df2.stack(future_stack=True)
stacked

# Las columnas pasan a tener relacion con los indices resultando asi que los valores de esas columnas se agrupen
# segun el tipo de medicamento y la anterior columna "A" o "B" clasificandose como en el resultado
```

```python
stacked.unstack()

# Con el comando .unstack todos los valores vuelven a como estaban antes del uso de .stack
```

```python
stacked.unstack(1)

# Con este comando se alternan lo que antes era el segundo indice a lo que ahora es el encabezado de las columnas
# y lo que antes eran las columnas del abecedario ahora forman parte del indice. 
```

```python
stacked.unstack(0)

# Ahora en este caso se toma el primer indice como encabezado y los numeros se mueven al primer indice. 
```

## Categoricos

```python
df = pd.DataFrame(
    {"id": [1, 2, 3, 4, 5, 6], "raw_grade": ["a", "b", "b", "a", "a", "e"]}
)
df["grade"] = df["raw_grade"].astype("category")
df["grade"]

# se crea un dataframe con datos categoricos, esot indica que los valores a b e son categorias asignadas al id.
```

```python
new_categories = ["very good", "good", "very bad"]
df["grade"] = df["grade"].cat.rename_categories(new_categories)

# Se crea una nueva variable donde se alamencas nuevos valores categoricos
```

```python
df["grade"] = df["grade"].cat.set_categories(
    ["very bad", "bad", "medium", "good", "very good"]
)

df["grade"]

# Agregamos esas nuevas categorias al df
```

```python
df.sort_values(by="grade")

# Ahora podemo sobservar como los valores categoricos de grande se alinean con los valores de raw_grade
# marcando asi si una de las letras del abecedario es buena o mala.
```

```python
df.groupby("grade", observed=False).size()

# Ahora se agurpan loc valores segun el tipo de categoria, contando la cantidad de valores por categoria.
```

## Plotting

```python
import matplotlib.pyplot as plt
plt.close("all")

# Importamos la libreria de matplotlib.pyplot, seguido de que se cierran todas las figuras 
# que están actualmente abiertas en la sesión de matplotlib
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))

ts = ts.cumsum()

ts.plot();

# Creamos un grafico con los valores de la variable ts donde se piden la creacion de 1000 valores aleatorios
# dentro de un periodo de 1000 dias apartir del 1/1/2000.
# cumsum() Calcula la suma acumulativa de los valores en la serie.
# .plot() Se usa para visualizar el grafico.
```

```python
df = pd.DataFrame(
    np.random.randn(1000, 4), index=ts.index, columns=["A", "B", "C", "D"]
)


df = df.cumsum()

plt.figure();
df.plot();

plt.legend(loc='best');

# Se crea una variable df donde se pide la creacion aleatoria de 1000 filas y 4 columnas las cuales tendran las letras del abecedario
# al igual que el anterior se calcula la suma acumulativa de los valores en la serie y luego de pide visualizar la data.
```
